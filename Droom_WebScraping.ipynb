```python
from bs4 import BeautifulSoup as soup
from selenium import webdriver
import requests
import pandas as pd
```

```python
# Url's
df=pd.read_excel('urls.xlsx',header=None)
url_list=df[0:401] # Tata Indica 202:301 224:235 275 275:401

# Data Lists
name=[] # List of Names
selling_price=[] # List of selling price
onroad_price=[] # List of On Road Price
# From - Full Circle Trust Score
Rating=[] 
Transparency_Score=[]
Seller_Score=[]
Health_Score=[]
Pricing_Score=[]
# For Basic Fact
make=[]
model=[]
trim=[]
year=[]
ext_color=[] # Interior Color
int_color=[] # Exterior Color
fuel_type=[]
trans_type=[] # Transmission Type
body_type=[]
int_furnishing=[] # Interior Furnishing
seating_capacity=[] # Seating Capacity
fuel_tank_capacity=[]
# Key Factors
location=[]
registration_state=[]
kilometers_driven=[]
seller_type_dealer=[]
num_of_owners=[]
# Technical Specification
engine_displacement=[]
fuel_efficiency=[]

# Primary Program
iteration=0
for link in url_list[0]:
    
    iteration=iteration+1
    print(iteration,link)
    
    # Chrome Driver
    driver=webdriver.Chrome()
    driver.get(link)
    source=driver.page_source
    page_soup=soup(source,'html.parser')
    
    # Name
    nam=(page_soup.find('div',{"class":"page-header","role":"global"}).text).replace('\n','').replace('Download Brochure','').strip()
    name.append(nam) # List of Names
    
    # SellingPrice
    selling_price.append((page_soup.findAll('div',{'class':'col-xl-4 col-lg-4 col-md-4 offer'}))[0].div.text.strip())
    
    # OnRoadlPrice
    op=page_soup.find('div',{"id":"dis-onroadprice"}).text
    onroad_price.append(op)
    
    # Score Rating
    ai=[]
    Score=page_soup.findAll('div',{"class":"score"})
    for i in Score:
        a=i.find("span").text
        ai.append(a)
    Rating.append(ai[0])
    Transparency_Score.append(ai[1])
    Seller_Score.append(ai[2])
    Health_Score.append(ai[3])
    Pricing_Score.append(ai[4])
    
    # For Panel Information
    tab_content=page_soup.findAll('div',{'class':'tab-content'})
    class_point=tab_content[0].findAll('div',{'class':'point'})
    make_bool=False
    model_bool=False
    trim_bool=False
    year_bool=False
    ext_color_bool=False
    int_color_bool=False
    fuel_type_bool=False
    trans_type_bool=False
    seating_capacity_bool=False
    body_type_bool=False
    int_furnishing_bool=False
    fuel_tank_capacity_bool=False
    location_bool=False
    kilometers_driven_bool=False
    seller_type_dealer_bool=False
    num_of_owners_bool=False
    engine_displacement_bool=False
    fuel_efficiency_bool=False
    for j in range(len(class_point)):
        if class_point[j].label.text.strip()=='Make':
            make.append(class_point[j].a.text.strip())
            make_bool=True
        if class_point[j].label.text.strip()=='Model':
            model.append(class_point[j].a.text.strip())
            model_bool=True
        if class_point[j].label.text.strip()=='Trim':
            trim.append(class_point[j].span.text.strip())
            trim_bool=True
        if class_point[j].label.text.strip()=='Year':
            year.append(class_point[j].span.text.strip())
            year_bool=True
        if class_point[j].label.text.strip()=='Exterior Color':
            ext_color.append(class_point[j].span.text.strip())
            ext_color_bool=True
        if class_point[j].label.text.strip()=='Interior Color':
            int_color.append(class_point[j].span.text.strip())
            int_color_bool=True
        if class_point[j].label.text.strip()=='Fuel Type':
            fuel_type.append(class_point[j].a.text.strip())
            fuel_type_bool=True
        if class_point[j].label.text.strip()=='Transmission Type':
            trans_type.append(class_point[j].a.text.strip())
            trans_type_bool=True
        if class_point[j].label.text.strip()=='Body Type':
            body_type.append(class_point[j].a.text.strip())
            body_type_bool=True
        if class_point[j].label.text.strip()=='Interior furnishing':
            int_furnishing.append(class_point[j].span.text.strip())
            int_furnishing_bool=True
        if class_point[j].label.text.strip()=='Seating Capacity':
            seating_capacity.append(class_point[j].span.text.strip())
            seating_capacity_bool=True
        
        if class_point[j].label.text.strip()=='Fuel Tank Capacity (in litres)':
            fuel_tank_capacity.append(class_point[j].span.text.strip())
            fuel_tank_capacity_bool=True
        if class_point[j].label.text.strip()=='Location':
            location.append(class_point[j].a.text.strip())
            location_bool=True
        if class_point[j].label.text.strip()=='Registration State':
            registration_state.append(class_point[j].span.text.strip())
            registration_state_bool=True
        if class_point[j].label.text.strip()=='Kilometers Driven':
            kilometers_driven.append(class_point[j].span.text.strip())
            kilometers_driven_bool=True
        if class_point[j].label.text.strip()=='Seller Type':
            seller_type_dealer.append(class_point[j].span.text.strip())
            seller_type_dealer_bool=True
        if class_point[j].label.text.strip()=='Number of Owners':
            num_of_owners.append(class_point[j].span.text.strip())
            num_of_owners_bool=True
        if class_point[j].label.text.strip()=='Displacement':
            engine_displacement.append(class_point[j].span.text.strip())
            engine_displacement_bool=True  
        if class_point[j].label.text.strip()=='Fuel Efficiency':
            fuel_efficiency.append(class_point[j].span.text.strip())
            fuel_efficiency_bool=True
        if j==len(class_point)-1:
            if make_bool==False:
                make.append('NA')
            if model_bool==False:
                model.append('NA')
            if trim_bool==False:
                trim.append('NA')
            if year_bool==False:
                year.append('NA')
            if ext_color_bool==False:
                ext_color.append('NA')
            if int_color_bool==False:
                int_color.append('NA')
            if fuel_type_bool==False:
                fuel_type.append('NA')
            if trans_type_bool==False:
                trans_type.append('NA')
            if seating_capacity_bool==False:
                seating_capacity.append('NA')
            if body_type_bool==False:
                body_type.append('NA')
            if fuel_tank_capacity_bool==False:
                fuel_tank_capacity.append('NA')
            if location_bool==False:
                location.append('NA')
            if registration_state_bool==False:
                registration_state.append('NA')
            if kilometers_driven_bool==False:
                kilometers_driven.append('NA')
            if seller_type_dealer_bool==False:
                seller_type_dealer.append('NA')
            if num_of_owners_bool==False:
                num_of_owners.append('NA')
            if engine_displacement_bool==False:
                engine_displacement.append('NA')
            if fuel_efficiency_bool==False:
                fuel_efficiency.append('NA')                 
# For building Dataframe
a=pd.DataFrame(name)
b=pd.DataFrame(selling_price)
c=pd.DataFrame(onroad_price)
d=pd.DataFrame(Rating)
e=pd.DataFrame(Transparency_Score)
f=pd.DataFrame(Seller_Score)
g=pd.DataFrame(Health_Score)
h=pd.DataFrame(Pricing_Score)
i=pd.DataFrame(make)
j=pd.DataFrame(model)
k=pd.DataFrame(trim)
l=pd.DataFrame(year)
m=pd.DataFrame(ext_color)
n=pd.DataFrame(int_color)
o=pd.DataFrame(fuel_type)
p=pd.DataFrame(trans_type)
q=pd.DataFrame(body_type)
r=pd.DataFrame(int_furnishing)
s=pd.DataFrame(seating_capacity)
t=pd.DataFrame(fuel_tank_capacity)
u=pd.DataFrame(location)
v=pd.DataFrame(registration_state)
w=pd.DataFrame(kilometers_driven)
x=pd.DataFrame(seller_type_dealer)
y=pd.DataFrame(num_of_owners)
z=pd.DataFrame(engine_displacement)
a_1=pd.DataFrame(fuel_efficiency)


main_df=pd.concat([a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z,a_1],axis=1)
my_col=['Name_of_Vehical','Selling_Price','On_Road_Price','Rating','Transparency_Score','Seller_Score','Health_Score','Pricing_Score','Make','Model','Trim','Year','Exterior_Color','Interior_Color','Fuel_Type','Transmission_Type','Body_Type','Interior_Furnishing','Seating_Capacity','Fuel_Tank_Capacity','Location','Registration_State','Kilometers_Driven','Seller_Type','Number_of_Owners','Engine_Displacement','Fuel_Efficiency']
main_df.columns=my_col
main_df.to_csv("used_car_price_prediction_test_11.csv")
```
